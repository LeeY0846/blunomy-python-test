# Fit curves

After clustering the points into groups, the next steps are:

1. Find the plane on which each wire lies
2. Fit a catenary model to the wire in that plane
3. Transform the fitted model back into 3D space

For the first step, PCA already gives us the along-wire direction, denoted by $e_1$. Together with the global vertical direction $z_{world}$, we can define the local vertical direction $e_2$ as: $e_2=z_{world}-(z_{world}\cdot e_1)\cdot e_1$

For convenience, we use the cluster centroid as the origin, and then project the points onto the plane.

Next, we use least squares to fit the projected points to the 2D catenary model and estimate the parameters $v_0$, $u_0$, and $c$.

The 2D catenary model is:

$$
v(u) = v_0 + c \times (\cosh{(u-u_0)/c} - 1)
$$

The resulting 3D catenary model is:

$$
P(u) = origin + u e_1 + v(u) e_2
$$



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy.optimize import least_squares
from research import read_dataset, DIFFICULTIES, cluster_wires

def catenary(u, c, u0, v0):
    return v0 + c * (np.cosh((u - u0) / c) - 1.0)

def residuals(params, u, v):
    c, u0, v0 = params
    pred = v0 + c * (np.cosh((u-u0)/c) - 1.0)
    return pred - v

data = read_dataset(DIFFICULTIES[1])

wires = cluster_wires(data)

wire_0 = wires[6]

points = wire_0.to_numpy()

pca = PCA(n_components=3)
pca.fit(points)

# Get the along-wire direction
e1 = pca.components_[0]
e1 = e1 / np.linalg.norm(e1)

z_world = np.array([0.0, 0.0, 1.0])
e2 = z_world - np.dot(z_world, e1) * e1
e2 = e2 / np.linalg.norm(e2)

# Get the plan normal
n = np.cross(e1, e2)
n = n / np.linalg.norm(n)

# Project points onto the plane
origin = points.mean(axis = 0)

points_relative = points - origin
dist_normal = points_relative @ n

points_projected = points - np.outer(dist_normal, n)

rel_proj = points_projected - origin
u = rel_proj @ e1
v = rel_proj @ e2

c0 = max((u.max() - u.min()) / 4, 1e-3)
u0_0 = u[np.argmin(v)]
v0_0 = v.min()
x0 = np.array([c0, u0_0, v0_0])

result = least_squares(
    residuals,
    x0=x0,
    args=(u, v),
    bounds=([1e-6, u.min(), v.min() - 10],
            [1e6, u.max(), v.max() + 10]),
    loss="soft_l1"
)

c_fit, u0_fit, v0_fit = result.x

u_line = np.linspace(u.min(), u.max(), 400)
v_line = catenary(u_line, c_fit, u0_fit, v0_fit)

curve_3d = origin + np.outer(u_line, e1) + np.outer(v_line, e2)

ax = plt.figure().add_subplot(projection="3d")
ax.scatter(points[:,0], points[:,1], points[:,2], c = "blue", alpha = 0.1, s = 2)
ax.scatter(curve_3d[:,0], curve_3d[:,1], curve_3d[:,2], c = "red", s = 1)

ax.view_init(elev=45, azim=30)

plt.show()
```

```python
def fit_curve(points):
    pca = PCA(n_components=3)
    pca.fit(points)

    # Get the along-wire direction
    e1 = pca.components_[0]
    e1 = e1 / np.linalg.norm(e1)

    z_world = np.array([0.0, 0.0, 1.0])
    e2 = z_world - np.dot(z_world, e1) * e1
    e2 = e2 / np.linalg.norm(e2)

    # Get the plan normal
    n = np.cross(e1, e2)
    n = n / np.linalg.norm(n)

    # Project points onto the plane
    origin = points.mean(axis = 0)

    points_relative = points - origin
    dist_normal = points_relative @ n

    points_projected = points - np.outer(dist_normal, n)

    rel_proj = points_projected - origin
    u = rel_proj @ e1
    v = rel_proj @ e2

    c0 = max((u.max() - u.min()) / 4, 1e-3)
    u0_0 = u[np.argmin(v)]
    v0_0 = v.min()
    x0 = np.array([c0, u0_0, v0_0])

    result = least_squares(
        residuals,
        x0=x0,
        args=(u, v),
        bounds=([1e-6, u.min(), v.min() - 10],
                [1e6, u.max(), v.max() + 10]),
        loss="soft_l1"
    )

    c_fit, u0_fit, v0_fit = result.x

    u_start = u.min()
    u_end = u.max()

    return {"e1": e1, "e2": e2, "origin": origin, "c_fit": c_fit, "u0_fit": u0_fit, "v0_fit": v0_fit, "u_start": u_start, "u_end": u_end}

def plot_fitted_curve(ax, fit_result:dict):
    u_line = np.linspace(fit_result["u_start"], fit_result["u_end"], 400)
    v_line = catenary(u_line, fit_result["c_fit"], fit_result["u0_fit"], fit_result["v0_fit"])

    curve_3d = fit_result["origin"] + np.outer(u_line, fit_result["e1"]) + np.outer(v_line, fit_result["e2"])

    ax.scatter(curve_3d[:,0], curve_3d[:,1], curve_3d[:,2], c = "red", s = 1)


fig = plt.figure(figsize=((12,10)))

for i,difficulty in enumerate(DIFFICULTIES):
    ax = fig.add_subplot(2,2,i+1,projection="3d")
    ax.set_title(difficulty.capitalize())
    data = read_dataset(difficulty)
    wires = cluster_wires(data)

    for wire in wires:
        points = wire.to_numpy()
        ax.scatter(points[:,0], points[:,1], points[:,2], c = "blue", alpha = 0.1, s = 2)
        fit_result = fit_curve(points)
        
        plot_fitted_curve(ax, fit_result)

    ax.view_init(elev=30, azim=40)

plt.show()
```

The algorithm performs well across all wires, as shown in the figures above.

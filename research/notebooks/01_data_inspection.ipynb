# Data Inspection

The parquet files provided in the case study contains the XYZ coordinates of points of wires. These wires share similar directions.

In the easy, hard, and extrahard datasets, there are three wires. There are seven wires in the medium dataset. Points in the hard dataset are more sparse than the others.

The wires in the given datasets can be differentiated on the XY-plane, so that would be a good start point to test clustering methods.

```python
import matplotlib.pyplot as plt
import pandas as pd
from research import read_dataset, DIFFICULTIES

data = read_dataset("easy")
data.head()
```

```python
# Print the points in 3D
fig = plt.figure(figsize=(12,10))

for i, diff in enumerate(DIFFICULTIES):
    ax = fig.add_subplot(2,2,i+1,projection="3d")
    data = read_dataset(diff)
    ax.scatter(data["x"], data["y"], data["z"], s = 2)
    ax.set_title(diff.capitalize())
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")

plt.tight_layout()
plt.show()
```

```python
# Print the projections on the XY-plane
fig = plt.figure(figsize=(12,10))

for i, diff in enumerate(DIFFICULTIES):
    ax = fig.add_subplot(2,2,i+1)
    data = read_dataset(diff)
    ax.scatter(data["x"], data["y"], s = 2)
    ax.set_title(diff.capitalize())
    ax.set_xlabel("X")
    ax.set_ylabel("Y")

plt.tight_layout()
plt.show()
```
